### Problem 1: Number of Substrings of one string present in other string

**Example**:
1) str1 = "aab", str2 = "abcd". Output -> 4 ("a","a","b","ab")
2) str1 = "abcd", str2 = "dcba". Output -> 4 ("a","b","c","d")
3) str1 = "aab", str2 = "aaab". Output -> 6 ("a","a","b","aa", "ab", "aab")

Ref: https://www.geeksforgeeks.org/number-of-substrings-of-one-string-present-in-other/



```python
def count_substrings(str1, str2):
    """Function to return total number of substrings matching str1 and str2"""
    count = 0
    # Loop to move outer window once all substrings are computed
    for i in range(len(str1)):
        c = ""
        # Loop to include character by character starting from i
        for j in range(i, len(str1)):
            c += str1[j]

            if str2.find(c) != -1:
                count += 1
    return count
```

```python
count_substrings("aab","abcd")
```

```python
count_substrings("abcd","dcba")
```

```python
count_substrings("aab","aaab")
```

##### **Note**: Order of complexity is O(n^3) 

### Problem 2: Print all substring of a number without any conversion

**Example:**

Input: 123 -> Output: 1, 12, 123, 2, 23, 3

Ref: https://www.geeksforgeeks.org/print-all-substring-of-a-number-without-any-conversion/



```python
import math
```

```python
n = 123
print(math.log10(123))
print(10 **2)
print(n % 100)
```

```python
def get_substrings_num(num):
    """Function to print all substrings of a number without any conversion"""
    len = int(math.log10(num))
    d1 = d2 = int(math.pow(10, len))
    while num > 0:
        # Print the floor value after dividing by d2
        while d2 > 0:
            print(num // d2)
            d2 //= 10
        # Get the reminder for next loop
        num %= d1

        d1 //= 10
        d2 = d1
```

```python
get_substrings_num(123)
```

```python
get_substrings_num(12345)
```
